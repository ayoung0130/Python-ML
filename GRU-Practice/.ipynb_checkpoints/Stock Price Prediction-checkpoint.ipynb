# GRU 기반의 삼성전자 주가 예측
- yahoo finance에서 데이터 다운로드 후 정확도 향상을 위해 3일(3MA), 5일(5MA) 가격이평선(이동평균선) 추가

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
```

```python
# 판다스 데이터프레임
# yahoo finance 로부터 데이터 다운로드]

raw_df = pd.read_csv('./005930.KS_3MA_5MA.csv') 
raw_df.head()

# Adj Close -> 조정 종가(주식이 한 거래일 동안 거래된 마지막 가격)
```

```python
plt.title('SAMSUNG ELECTRONIC STCOK PRICE')
plt.ylabel('price')
plt.xlabel('period')
plt.grid()

plt.plot(raw_df['Adj Close'], label='Adj Close')

plt.show()
```

### 데이터 전처리 (Missing Data 처리, 정규화 등)

```python
# 통계 정보 확인
raw_df.describe()
```

```python
# Missing Data 확인
raw_df.isnull().sum()
```

```python
# 최소값이 0인 column 체크
for col in raw_df.columns:
    if raw_df[col].min() == 0:
        col_name = col
        print(col_name, type(col_name))
```

```python
# Volume 열이 0인 데이터를 추출
# .loc[]: 레이블 기반의 인덱싱을 수행하는 메소드
raw_df.loc[raw_df['Volume']==0]
```

```python
# 각 column에 0이 몇개인지 확인
for col in raw_df.columns:
    missing_rows = raw_df.loc[raw_df[col]==0].shape[0]
    print(col + ': ' + str(missing_rows))
```

```python
# 0을 NaN으로 바꾼 후 Missing Data 처리
raw_df['Volume'] = raw_df['Volume'].replace(0, np.nan)

# 각 column에 0이 몇개인지 확인
for col in raw_df.columns:
    missing_rows = raw_df.loc[raw_df[col]==0].shape[0]
    print(col + ': ' + str(missing_rows))
```

```python
# Missing Data 확인
raw_df.isnull().sum()
```

```python
raw_df.isnull().any()
```

```python
# 'Open' 열의 값이 NaN인 데이터를 추출
raw_df.loc[raw_df['Open'].isna()]
```

```python
# Missing Data 처리
# dropna() -> NaN 존재하는 행을 제거
raw_df = raw_df.dropna()
raw_df.isnull().sum()
```

```python
# 정규화 (Data를 제외한 모든 수치부분을 정규화)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# 정규화할 열 선택
scale_cols = ['Open', 'High', 'Low', 'Close', 'Adj Close',
              '3MA', '5MA', 'Volume']

scaled_df = scaler.fit_transform(raw_df[scale_cols]) # 정규화

scaled_df = pd.DataFrame(scaled_df, columns=scale_cols) # 데이터프레임으로 변환

print(scaled_df)
```

### 주가 예측을 위해 3MA, 5MA, Adj Close 항목을 feature 선정
- 정답은 Adj Close 선정
- 시계열 데이터를 위한 window_size = 40 선정
- window_size : 시퀀스 데이터의 윈도우 크기(=각 시퀀스 데이터의 길이)

```python
# 입력 파라미터 feature, 출력 파라미터 label -> numpy type

def make_sequene_dataset(feature, label, window_size):
    feature_list = [] # 생성될 feature list
    label_list = []   # 생성될 label list
    
    for i in range(len(feature)-window_size):
        feature_list.append(feature[i:i+window_size])
        label_list.append(label[i+window_size])
        
    return np.array(feature_list), np.array(label_list)
```

```python
# feature_df, label_df 생성

feature_cols = [ '3MA', '5MA', 'Adj Close' ]
label_cols = [ 'Adj Close' ]

feature_df = pd.DataFrame(scaled_df, columns=feature_cols)
label_df = pd.DataFrame(scaled_df, columns=label_cols)
```

```python
# DataFrame을 numpy로 변환

feature_np = feature_df.to_numpy()
label_np = label_df.to_numpy()

print(feature_np.shape, label_np.shape)
```

### 시계열 데이터 생성 (make_sequence_dataset)

```python
window_size = 40

X, Y = make_sequene_dataset(feature_np, label_np, window_size)

print(X.shape, Y.shape)
```

### train data, test data 생성

```python
# train, test 분리

split = -200

x_train = X[0:split]
y_train = Y[0:split]

x_test = X[split:]
y_test = Y[split:]

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

### 모델 구축 및 컴파일

```python
# model 생성
# Sequential 모델: 레이어를 선형으로 쌓아 구성하는 가장 간단한 형태의 신경망 모델
model = Sequential()

# GRU 레이어 추가 (뉴런 개수, 활성화 함수, 입력 데이터의 모양)
# 입력 데이터 모양 -> x_train의 첫 번째 데이터 샘플의 모양을 사용
model.add(GRU(256, activation='tanh',input_shape=x_train[0].shape))

# Dense 레이어 추가
# Dense: 은닉층, 완전 연결(fully connected) 레이어.
model.add(Dense(1, activation='linear'))

# 모델 요약 정보 출력
model.summary()
```

```python
# 손실 함수 -> mse(Mean Squared Error) - 평균 제곱 오차
# 옵티마이저 -> adam
# 평가 지표 -> mae(Mean Absolute Error) - 평균 절대 오차 사용
model.compile(loss='mse', optimizer='adam', metrics=['mae'])

model.summary()
```

### 모델 학습 (EarlyStopping 적용)

```python
# 학습을 조기 종료할 수 있도록 해주는 클래스 불러오기
from tensorflow.keras.callbacks import EarlyStopping

# EarlyStopping(모니터링할 지표, 지표가 개선되지 않아도 몇 번까지 기다릴 것인지)
early_stop = EarlyStopping(monitor='val_loss', patience=5)

model.fit(x_train,y_train,
          validation_data=(x_test, y_test),
          epochs=100, batch_size=16,
          callbacks=[early_stop])
```

### 예측을 통한 정답과의 비교. 오차계산-MAPE(평균절대값백분율오차)

```python
pred = model.predict(x_test)

plt.figure(figsize=(12,6))
plt.title('Stock Price Prediction')
plt.ylabel('adj close')
plt.xlabel('period')
plt.plot(y_test, label='actual')
plt.plot(pred, label='prediction')
plt.grid()
plt.legend(loc='best')  # 범례 표시, 위치는 알아서 최적의 위치로 설정

plt.show()
```

```python
# MAPE(평균절대값백분율오차계산) -> 값이 작을수록 정확

print(np.sum(abs(y_test-pred)/y_test)/len(x_test))  # 결과: 2.15%
```
